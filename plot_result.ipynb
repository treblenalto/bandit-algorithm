```python
import plotly.express as px
import pandas as pd

df = pd.read_csv("results.csv")

offset = 0
for algo in ["Greedy", "EpsilonGreedy", "UCB", "ThompsonSampling"]:
    df.loc[df["algorithm"] == algo, "E(P(A))"] += offset
    df.loc[df["algorithm"] == algo, "E(P(B))"] += offset
    offset += 0.01

fig = px.scatter(df, x="E(P(A))", y="E(P(B))", size="score", color="algorithm")
fig.update_traces(marker=dict(line=dict(width=0)), opacity=0.4)
fig.update_layout(
    title="Multi-Armed Bandit", xaxis_title="E(P(A))", yaxis_title="E(P(B))"
)
fig.show()
```

```python
df.groupby("algorithm").mean()["score"].sort_values(ascending=False)
```
